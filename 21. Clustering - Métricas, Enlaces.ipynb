# Distancias

```python
from scipy.spatial import distance_matrix
import pandas as pd
```

```python
data = pd.read_csv("../Datasets/movies/movies.csv", sep=";")
data
```

```python
movies = data.columns.values.tolist()[1:]
movies
```

Para hacer clustering se necesita una matriz de distancias entre los individuos del dataset, puede ser distancia euclidea, manhatan, minkowski, etc

```python
dd1 = distance_matrix(data[movies], data[movies], p=1) #obtiene matriz de distancias de manhatan
dd2 = distance_matrix(data[movies], data[movies], p=2)  #obtiene matriz de distancias euclideas
dd10 = distance_matrix(data[movies], data[movies], p=10) #obtienes matriz de distancias de minkowsky con p=10
```

```python
#Hacer funcion para mostrar el array que devuelve distance_matrix como un dataframe
def dm_to_df(dd, col_name):     
    import pandas as pd
    return pd.DataFrame(dd, index=col_name, columns=col_name)
```

```python
dm_to_df(dd1, data["user_id"])
```

```python
dm_to_df(dd2, data["user_id"])
```

```python
dm_to_df(dd10, data["user_id"])
```

# Enlaces

Para comenzar a clusterizar, se calculan distancias entre clusters, la primera matriz de distancias entre clusters es la matriz de distancias entre individuos, se toma cada individuo como si fuera un cluster de un solo elemento. 

Luego se elige la distancia ms pequea entre dos clusters y se aplica algn mtodo de enlace, simple, promedio, completo, etc para obtener la nueva distancia entre el nuevo cluster y los dems clusters. Los enlaces sirven para definir las nuevas distancias entre el cluster generado y los demas. 

Los enlaces dicen de qu manera calcular distancias entre clusters

```python
df = dm_to_df(dd1, data["user_id"])
df
```

```python
Z=[] #Matriz de enlaces: Cada renglon de esta matriz indica el enlace para la creacion de un nuevo cluster.
# columnas: [cluster1, cluster2, distancia a la que se unen, elementos contenidos en la union cluster1+cluster2]
```

```python
df[11]=df[1]+df[10]  #Crea nueva columna con el mismo tipo de dato que las otras columnas
df.loc[11]=df.loc[1]+df.loc[10] #Crea unevo renglon con el mismo tipo de dato que los otros renglones
Z.append([1,10,0.7,2])#id1, id2, d, n_elementos_en_cluster -> cluster 11.
df
```

```python
#Liga simple: distancia entre dos clusters=minimo de las distancias entre los elementos de cada uno de los clusters
#Aqui se rellena la columna 11 y el renglon 11 con la liga simple
for i in df.columns.values.tolist():
    df.loc[11][i] = min(df.loc[1][i], df.loc[10][i])
    df.loc[i][11] = min(df.loc[i][1], df.loc[i][10])
df
```

```python
#Se borran renglon 1 y 10 y columna 1 y 10 
#porque ahora forman el cluster 11 representado por la columna 11 y renglon 11
df = df.drop([1,10]) 
df = df.drop([1,10], axis=1)
df
```

```python
#Se repite el proceso de union anterior
#Ahora se unen cluster2 y cluster7 formando el cluster12
x = 2
y = 7

n = 12

#Se crea renglon 12 y columna 12 del mismo tamao y tipo
df[n]=df[x]+df[y]
df.loc[n]=df.loc[x]+df.loc[y]

#Se rellena siguiente renglon de la matriz de enlaces
Z.append([x,y,df.loc[x][y],2])#id1, id2, d, n_elementos_en_cluster -> 11.

#Se rellenan renglon 12 y columna 12 con los valores del enlace simple
for i in df.columns.values.tolist():
    df.loc[n][i] = min(df.loc[x][i], df.loc[y][i])
    df.loc[i][n] = min(df.loc[i][x], df.loc[i][y])

#Se borran clusters 2 y 7 de la matriz de distancias
df = df.drop([x,y])
df = df.drop([x,y], axis=1)
df
```

# Clustering jerrquico

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
movies
```

```python
data[movies]
```

```python
#Funcion linkage crea automaticamente la matriz de enlaces y la devuelve. 
# linkage[dataset de individuos, tipo de enlace, metrica(euclidea predeterminada)]
Z = linkage(data[movies], "ward") 
Z
plt.figure(figsize=(25,10))
plt.title("Dendrograma jerrquico para el Clustering")
plt.xlabel("ID de los usuarios de Netflix")
plt.ylabel("Distancia")
#Funcion dendogram devuelve un diccionario que contiene en cada definicion 
#la informacion acerca de una variable necesaria para la grfica del dendograma
dendrogram(Z, leaf_rotation=90., leaf_font_size=10.0) 
plt.show()
```

```python
Z = linkage(data[movies], "average")
Z
plt.figure(figsize=(25,10))
plt.title("Dendrograma jerrquico para el Clustering")
plt.xlabel("ID de los usuarios de Netflix")
plt.ylabel("Distancia")
dendrogram(Z, leaf_rotation=90., leaf_font_size=10.0)
plt.show()
```

```python
data[movies]
```

```python
Z = linkage(data[movies], "complete")
Z
plt.figure(figsize=(25,10))
plt.title("Dendrograma jerrquico para el Clustering")
plt.xlabel("ID de los usuarios de Netflix")
plt.ylabel("Distancia")
dendrogram(Z, leaf_rotation=90., leaf_font_size=10.0)
plt.show()
```

```python
Z = linkage(data[movies], method="single", metric="cosine")
Z
plt.figure(figsize=(25,10))
plt.title("Dendrograma jerrquico para el Clustering")
plt.xlabel("ID de los usuarios de Netflix")
plt.ylabel("Distancia")
dendrogram(Z, leaf_rotation=90., leaf_font_size=10.0)
plt.show()
```
