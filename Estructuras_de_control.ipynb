```python
# IF 
#se difine mediante indentacion, es decir, espaciado para cada bloque

#if simple
if 2<3:
    print("2 es menor que 3")
    
#if-else   
if 2>3:
    print("2 es mayor que 3")
else:
    print("2 no es mayor que 3")
    
#if-elif-else
if 3==8:
    print("3 es igual a 8")
elif 3<8:
    print("3 es menor que 8")
else:
    print("3 es mayor que 8")
    
#if de un solo renglon
if "hola"=="hola": print("hola y hola son iguales")
    
#if-else de un solo renglon
print("5 es mayor a 6") if 5>6 else print("5 no es mayor a 6")

#if-elif-else de un solo renglon
print("2 > 3") if 2>3 else print("5 > 5") if 5>5 else print("0 > 1") if 0>1 else print("ninguna de las anteriores")

#if anidado
x="hola"
y="hola caracol"

if x in y:
    if x==y:
        print("son iguales")
    else:
        print("aunque",x,"esta contenido en", y,"no son iguales")
else:
    print(x,"no está en",y)

# bloque vacío
if 2==3:
    pass       # se debe de señalar pass cuando un bloque no tiene contenido, para que no mande error
print("aquí sigue el flujo del codigo")

#contraejemplo
#if 2==3:
```

```python
# WHILE

#while simple
i=1
while i<6:
    print(i)
    i+=1

# break - para salirdel loop auque la condicion de while sea verdadera
print()
i=1
while i<6:
    print(i)
    if i>4:
        break
    i+=1
    
# continue - para detener la iteracion actual y volver a evaluar en el while
print()
i=1
while i<10:
    i+=1
    if i==5:
        continue    #cuando i==5 saldrá del loop (por lo tanto ya no imprimirá el 5) y regresará a la evaluacion del while
    print(i)

# else - para añadir una sentencia cuando deje de cumplirse el while
print()
i=1
while i<5:
    print(i)
    i+=1
else:
    print("Ahora i=",i)
```

```python
# FOR
#es un ciclo que se hace sobre un arreglo (ya sea String, lista, tupla, diccionario, conjunto, range)

#for simple
for x in (1,2,3):
    print("tupla",x)

for x in [1,2,3]:
    print("lista",x)

for x in "CADENA":
    print(x)
    
for x in range(5):
    print("  ",x)

for x in range(10,20,2):
    print(x)
    
#break - sale del ciclo totalmente
print()
for x in "CADENA":
    if x =="E":
        break
    print(x)

#continue - detiene la iteracion actual pero regresa a evaluar el for
print()
for x in "CADENA":
    if x =="E":
        continue
    print(x)

#else - para agregar una sentencia cuando haya terminado de recorrerse el arreglo
print()
for x in (1,2,3):
    print("tupla",x)
else:
    print("se terminó la tupla")
    
#pass - se debe de escribir en un bloque vacío para que no mande error
for i in ("a","b","c"):
    pass
```
